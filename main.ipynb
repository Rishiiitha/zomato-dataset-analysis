```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
#to join two dataset
final_df=pd.merge(df,df_country,on='Country Code',how="left")
final_df.head()
```

```python
## to check datatypes
final_df.dtypes
```

```python
countryname=final_df["Country"].value_counts().index
```

```python
countrybranch=final_df["Country"].value_counts().values
```

```python
plt.pie(countrybranch[:3],labels=countryname[:3],autopct="%1.2f%%")
```

Observation:
      Zomato larger transection is from India,then United states and then United Kingdom

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

#observation
1.when the rating is between 4.5 to 4.9 --->excellent
2.When rating is between 4.0 to 3.4-->very good
3.when the rating is between 3.5 to 3.9-->good
4.When Rating is between 2.5to 3.4   --->average
5.when the rating is between 1.8 to 2.4 -->poor

```python
plt.pie(ratings["Rating Count"],labels=ratings["Aggregate rating"])
```

```python

sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```
